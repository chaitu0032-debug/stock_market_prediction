```python
import pandas as pd
```

```python
df1 = pd.read_csv("ADANIPORTS.csv")
```

```python
df2 = pd.read_csv('compressed_2.csv')
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
df1
```

```python
df2['Date'] = pd.to_datetime(df2['Date'])
```

```python
df2
```

```python
 df3 = df2[::-1].reset_index()
```

```python
df4 = df3.drop(["index","Unnamed: 0"],axis = 1)
```

```python
df3 = df4
```

```python
df3
```

```python
df3["High"] = -100
df3["Low"] = -100
df3["VWAP"] = -100
df3["Turnover"] = -100
```

```python
df3
```

```python
for index in range(len(df3)):
    date = df3.loc[index,"Date"]
    cor_idx = df1.index[df1["Date"] == date]
    if(len(cor_idx) != 0):
        df3.loc[index,"High"] = df1.loc[cor_idx[0],"High"]
        df3.loc[index,"Low"] = df1.loc[cor_idx[0],"Low"]
        df3.loc[index,"VWAP"] = df1.loc[cor_idx[0],"VWAP"]
        df3.loc[index,"Turnover"] = df1.loc[cor_idx[0],"Turnover"]
```

```python
indexNames = df3[(df3['High'] == -100) ].index
```

```python
df3.drop(indexNames , inplace=True)
```
